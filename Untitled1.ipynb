```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="/content/drive/MyDrive/Tugas PSSO/data0-SZL.csv"
df= pd.read_csv(path)
df.head()
```

```python
mean_values = df.mean()
print(mean_values)
```

```python
std_values = df.std()
print(std_values)
```

```python
skew_values = df.skew()
print(skew_values)
```

```python
median_values = df.median()
print(median_values)
```

```python
energy_values = np.sum(df**2)
print(energy_values)
```

```python
min_values = df.min()
print(min_values)
```

```python
max_values = df.max()
print(max_values)
```

```python
statistics_df = pd.DataFrame({
    'Mean': mean_values,
    'Std Dev': std_values,
    'Skewness': skew_values,
    'Median': median_values,
    'Energy': energy_values,
    'Min': min_values,
    'Max': max_values
})
print(statistics_df)
```
